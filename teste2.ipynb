```python
J = [rnd.uniform(-10,10) for i in range (250)]
print(J)
```

COMO EXPORTAR A LISTA PARA FICHEIRO .CSV

```python
with open('unordered.csv', mode='w') as file:
    writer = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    writer.writerow(J)
```
